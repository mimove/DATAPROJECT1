```python
import os
import json
import geopandas as gpd
import numpy
import psycopg2
import random
```

```python
path = '../../datos/datos_out/barrios_updated.geojson'

with open('../../datos/datos_out/barrios_updated.geojson') as json_file:
    json_data = json.load(json_file)

barrios_json = []
for i in range(len(json_data['features'])):
    barrios_json.append(json_data['features'][i]) 
    
barrios_gpd = gpd.GeoDataFrame.from_features(barrios_json)
barrios_gpd.crs = 'epsg:4326'
```

```python
barrios_gpd
```

```python
barrios_gpd.drop(['last_edited_user'], axis = 1)
barrios_gpd
```

```python
random_quality = ['Desfavorable', 'Regular', 'Razonablemente buena', 'Buena']


for i in barrios_gpd['calidad_ambiental']:
    
    barrios_gpd['calidad_ambiental'] = random.choices(random_quality, k = len(barrios_gpd['calidad_ambiental']))
        
print(barrios_gpd['calidad_ambiental'])
```

```python
score = [1,2,3,4]
punt_ambiental = []

for i in barrios_gpd['calidad_ambiental']:
    
    if i == 'Desfavorable':
        punt_ambiental.append(score[0])

    elif i == 'Regular':
        punt_ambiental.append(score[1])

    elif i == 'Razonablemente buena':
        punt_ambiental.append(score[2])

    elif i == 'Buena':
        punt_ambiental.append(score[3])

print(*punt_ambiental)

barrios_gpd['punt_ambiental'] = punt_ambiental

print(barrios_gpd[['calidad_ambiental', 'punt_ambiental']])
```
